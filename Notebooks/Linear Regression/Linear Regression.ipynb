```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(formatter={'float': lambda x: "{0:0.6f}".format(x)})
import seaborn as sns
sns.set()
```

```python
salary_data = pd.read_csv("Salary_Data.csv")
x = salary_data["YearsExperience"].values.reshape(-1,1)
y = salary_data["Salary"].values.reshape(-1,1)
x = x / max(x)
y = y / max(y)

#Plotting data
plt.scatter(x,y,c='red')
plt.title("Years of experience vs Salary")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.show()
```

# Linear Regression Formula

The formula for calculating the weight matrix \( \mathbf{W} \) in a linear regression model, based on the input features \( \mathbf{X} \) and output labels \( \mathbf{Y} \), is expressed as follows:

\[
\mathbf{W} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{Y}
\]

Where:

- \( \mathbf{W} \) is the weight matrix we want to calculate.
- \( \mathbf{X} \) is a matrix containing the input features. Each row represents an instance, and each column represents a feature. The matrix \( \mathbf{X} \) has dimensions \( m \times n \), where \( m \) is the number of instances, and \( n \) is the number of features.
- \( \mathbf{Y} \) is a vector of output labels.

This formula is commonly used in linear regression to estimate the optimal weights that minimize the mean squared error (MSE) between the model's predictions and the actual \( \mathbf{Y} \) values.
