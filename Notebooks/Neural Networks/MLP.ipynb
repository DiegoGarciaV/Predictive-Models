```python
class MultilayerPerceptron:
    
    #Coleccion de numero de neuronas por capa
    T = []
    
    #coleccion de transformaciones lineales
    F = []
    L = 0
    
    #Coleccion de umbrales 
    B = []
    
    #Coleccion de salidas de cada capa (Se utilizarán para el Backpropagation)
    
    A = []
    
    def __init__(self,F0,B0,init_learning_rate=0.01):
        
        self.F = F0
        L = len(F0)
        self.B = B0
        self.init_learning_rate= init_learning_rate;
        self.learning_rate= init_learning_rate;
        
    def __init__(self,Top,init_learning_rate=0.01):
        
        L = len(Top)
        self.F = []
        self.B = []
        self.T = Top
        self.init_learning_rate= init_learning_rate;
        self.learning_rate= init_learning_rate;
        for t in range(0,len(Top)-1):
            
            self.B.append(np.random.rand(1,Top[t+1])* 4 -2 ) 
            self.F.append(np.random.rand(Top[t],Top[t+1])*4 -2)
            
            
            
        
    #funcion de activación
    def sigmoid(self,x):
        
            sig = 1 / (1 + np.exp(-x))
            return sig
        
    def sigmoid_derivative(self, x):
        return x * (1 - x)
    
    def forward(self,X):
        
        Fi = self.F
        Bi = self.B
        self.A = []
        
        if(len(Fi) != len(Bi)):
            return "Matrices incompatibles"
        
        Xi = X

        
        for i in range(0,len(Fi)):
            
            Ti = Fi[i]
            Y =  Xi @ Ti + Bi[i]
            
            Xi = self.sigmoid(Y)
            self.A.append(Xi)
            
            
        return np.array(Xi)
    

    def neuronsCount(self):
        
        k = 0
        for Fi in self.F:
            k = k + len(Fi)*(len(Fi[0])+1)
            
        return k
    
    
    def backward(self,X,Y,cost,learning_rate):

        a = self.forward(X)

        #backProp

        deltas = []

        for i in reversed(range(0,len(self.F))):

            a = self.A[i]
            if(i == len(self.F)-1):
                deltas.insert(0,cost(a,Y) * (a*(1-a)))
            else:
                deltas.insert(0,deltas[0] @ Fi * (a*(1-a)))

            Fi = self.F[i].T

            self.B[i] = self.B[i] - np.sum(deltas[0],axis=0,keepdims=True) * learning_rate

            if(i>0):
                ai = self.A[i-1].T
            else:
                ai = X.T
            self.F[i] = self.F[i] - ai  @ deltas[0] * learning_rate
    
        return  self.error(self.forward(X),Y)

    
    def error(self,X,Y):
    
        return np.mean((X-Y)**2)
        
    def gError(self,X,Y):
        
        return (X-Y)


    def train(self, X, Y, epochs=1000):

        error_min = 1;
        self.learning_rate = self.init_learning_rate;
        for i in range(0,epochs):
    
            valuation = self.backward(X,Y,self.gError,self.learning_rate)
            epoch_error = self.error(valuation,Y)

        print("Training finished")

        
```
